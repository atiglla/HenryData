```python
import os
archivo = open(r'C:\Users\Alfredo\Bitácoras\SoyHenry\DS-M1\Clase 04\hospitales2.csv')
dicc = {'WKT': [],
        'ID': [],
        'NOMBRE': [],
        'NOM_MAP': [],
        'OBJETO': [],
        'CALLE': [],
        'ALTURA': [],
        'DOM_NORMA': [],
        'DOM_GEO': [],
        'TELEFONO': [],
        'GUARDIA': [],
        'FAX': [],
        'WEB': [],
        'COD_POSTAL': [],
        'TIPO': [],
        'TIPO_ESPEC': [],
        'MOD_AT_1': [],
        'MOD_AT_2': [],
        'DEPEND_ADM': [],
        'DIRECTOR': []}
```

```python
next(archivo)
for linea in archivo: 
    linea = linea.strip()
    campos = linea.split(',')
    dicc['WKT'].append(campos[0])
    dicc['ID'].append(campos[1])
    dicc['NOMBRE'].append(campos[2])
    dicc['NOM_MAP'].append(campos[3])
    dicc['OBJETO'].append(campos[4])
    dicc['CALLE'].append(campos[5])
    dicc['ALTURA'].append(campos[6])
    dicc['DOM_NORMA'].append(campos[7])
    dicc['DOM_GEO'].append(campos[8])
    dicc['TELEFONO'].append(campos[9])
    dicc['GUARDIA'].append(campos[10])
    dicc['FAX'].append(campos[11])
    dicc['WEB'].append(campos[12])
    dicc['COD_POSTAL'].append(campos[13])
    dicc['TIPO'].append(campos[14])
    dicc['TIPO_ESPEC'].append(campos[15])
    dicc['MOD_AT_1'].append(campos[16])
    dicc['MOD_AT_2'].append(campos[17])
    dicc['DEPEND_ADM'].append(campos[18])
    dicc['DIRECTOR'].append(campos[19])
```

```python
'''Ahora un poco de data wrangling ó ETL por así decirlo:
Le sacamos el "POINT a la lista de valores, y la otra comilla " al final.
Y guardamos los nuevos valores corregidos en una lista nueva, llamada coordenadas.'''
coordenadas = []
for elemento in (dicc['WKT']):
    elemento = elemento.replace('"POINT ','')
    elemento = elemento.replace('"','')
    elemento = elemento.replace('(', '')
    elemento = elemento.replace(')', '')
    coordenadas.append(elemento)
        
print (coordenadas)
```

```python
'''De repente se me cayó una idea de hacer lo más simple y fácil. Decirle a Python que iguale la lista
WKT del diccionario a a lista que creamos recién.
Y al parecer, funcionó 👏👏👏'''
dicc['WKT'] = coordenadas
dicc['WKT']
dicc
```

```python
'''Ahora deberíamos separar esas coordenadas en dos columnas distintas. Latitud y Longitud.
Pero en inglés, tal como lo pide el ejercicio.
'''
LATITUDE = []
LONGITUDE = []
for index, elemento in enumerate(dicc['WKT']):
    elemento = elemento.split(' ')
    LATITUDE.append(elemento[0])
    LONGITUDE.append(elemento[1])
print(LATITUDE)
print(LONGITUDE)
```

```python
'''Agregamos la nueva columna al diccionario latitud y longitud'''
dicc['LATITUDE'] = LATITUDE
dicc['LONGITUDE'] = LONGITUDE
```
